```python
test = 'components/personas/empathetic_guide.txt'
test = os.path.dirname(test)
actual_path_test = f'../{test}'
# check path exists
if os.path.exists(actual_path_test):
    print(f"Path exists: {actual_path_test}")

df['real_path'] = df['output_path'].apply(lambda x: f'../{x}')

def save_file(content, real_path):
    with open(real_path, encoding='utf-8', mode='w') as f:
        f.write(content)
    
    # check file exists in real_path:
    assert os.path.exists(real_path), f"File does not exist at {real_path}"
    return True

df['COMPONENT_CREATED'] = df.apply(lambda x: save_file(
    content=x['RESULT'], 
    real_path=x['real_path'] if x['output_path'] else None
), axis=1)

df
```
